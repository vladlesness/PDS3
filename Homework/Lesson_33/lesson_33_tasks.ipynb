```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('car details v4.csv')
df
```

```python
fig = plt.figure(figsize=(8, 6), dpi=120)
ax = fig.add_subplot()
plt.plot(df.sort_values(by='Year')['Year'][0:30], df.sort_values(by='Year')['Price'][0:30])
```

```python
df = df.astype({'Fuel Type': str, 'Engine': str})
plt.scatter(df['Fuel Type'][0:50], df['Engine'][0:50])
```

```python
plt.bar(df['Year'], df['Kilometer'])
```

```python
plt.hist(df['Year'], 5, width=4)
```

```python
sb.boxplot(x='Owner', y='Price', data=df)
```

```python
sb.pairplot(df)
```
