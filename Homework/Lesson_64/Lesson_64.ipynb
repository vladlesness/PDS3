```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
chess1, chess2 = cv2.imread('chess_1.png'), cv2.imread('chess_2.png')
```

```python
fig, axis = plt.subplots(1, 2)
axis[0].imshow(chess1)
axis[1].imshow(chess2)
```

```python
chess1_corners = np.array([[192, 49], [463, 38], [198, 244], [458, 261]])
chess2_corners = np.array([[188, 316], [165, 34], [407, 363], [530, 102]])
```

```python
h, status = cv2.findHomography(chess2_corners, chess1_corners)
```

```python
warp_img = cv2.warpPerspective(chess2, h, (chess2.shape[1], chess2.shape[0]))
```

```python
plt.imshow(warp_img)
```
