Колонки Age, Cabin та Embarked мають пропущені значення.
Оскільки Age має дискретні кількісні значення то пропуски можна замінити середнім значенням.
Cabin та Embarked не несуть ніякого статистичного навантаження, тому їх можна не використовувати в навчанні моделі.

```python
df['Age'].fillna(df['Age'].mean().round(), inplace=True)
df.info()
```

```python
sns.displot(df['Age'], kde=True, bins=50)
```

```python
sns.displot(df, x='Age', kde=True, hue='Sex', bins=50)
```

```python
sns.displot(df, x='Age', kde=True, hue='Survived', bins=50)
```

```python
sns.displot(df, x='Age', kde=True, hue='Pclass', bins=50, palette=['green', 'red', 'lightblue'])
```
