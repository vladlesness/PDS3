```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels.api import qqplot
from sklearn.preprocessing import PowerTransformer
from scipy.stats import shapiro
from numpy import log, cbrt
```

```python
df = pd.read_csv('Data_to_Transform.csv')
df.head()
```

```python
df.info()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Initial distributions')
sns.histplot(data=df.iloc[:, 0], ax=ax[0, 0])
sns.histplot(data=df.iloc[:, 1], ax=ax[0, 1])
sns.histplot(data=df.iloc[:, 2], ax=ax[1, 0])
sns.histplot(data=df.iloc[:, 3], ax=ax[1, 1])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Initial distributions')
qqplot(df.iloc[:, 0], ax=ax[0, 0], line='45')
qqplot(df.iloc[:, 1], ax=ax[0, 1], line='45')
qqplot(df.iloc[:, 2], ax=ax[1, 0], line='45')
qqplot(df.iloc[:, 3], ax=ax[1, 1], line='45')
plt.show()
```

```python
print('Shapiro-Wilk Test for Gaussian distribution on initial data\n')
for col in df:
    stat, p = shapiro(df[col])
    print(col, '\nStatistic value:', stat, '\nP-value:', p, '\n')
```

```python
df_log = log(df)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distributions after logarithmic transformation')
sns.histplot(data=df_log.iloc[:, 0], ax=ax[0, 0])
sns.histplot(data=df_log.iloc[:, 1], ax=ax[0, 1])
sns.histplot(data=df_log.iloc[:, 2], ax=ax[1, 0])
sns.histplot(data=df_log.iloc[:, 3], ax=ax[1, 1])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distributions after logarithmic transformation')
qqplot(df_log.iloc[:, 0], ax=ax[0, 0], line='45')
qqplot(df_log.iloc[:, 1], ax=ax[0, 1], line='45')
qqplot(df_log.iloc[:, 2], ax=ax[1, 0], line='45')
qqplot(df_log.iloc[:, 3], ax=ax[1, 1], line='45')
plt.show()
```

```python
print('Shapiro-Wilk Test for Gaussian distribution after logarithmic transformation\n')
for col in df_log:
    stat, p = shapiro(df_log[col])
    print(col, '\nStatistic value:', stat, '\nP-value:', p, '\n')
```

```python
df_cbrt = cbrt(df)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distributions after cubit root transformation')
sns.histplot(data=df_cbrt.iloc[:, 0], ax=ax[0, 0])
sns.histplot(data=df_cbrt.iloc[:, 1], ax=ax[0, 1])
sns.histplot(data=df_cbrt.iloc[:, 2], ax=ax[1, 0])
sns.histplot(data=df_cbrt.iloc[:, 3], ax=ax[1, 1])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distributions after cubit root transformation')
qqplot(df_cbrt.iloc[:, 0], ax=ax[0, 0], line='45')
qqplot(df_cbrt.iloc[:, 1], ax=ax[0, 1], line='45')
qqplot(df_cbrt.iloc[:, 2], ax=ax[1, 0], line='45')
qqplot(df_cbrt.iloc[:, 3], ax=ax[1, 1], line='45')
plt.show()
```

```python
print('Shapiro-Wilk Test for Gaussian distribution after cubit root transformation\n')
for col in df_cbrt:
    stat, p = shapiro(df_cbrt[col])
    print(col, '\nStatistic value:', stat, '\nP-value:', p, '\n')
```

```python
df_boxcox_yeo_johnson = pd.DataFrame(columns=df.columns)

for i in range(4):
    if (df.iloc[:, i] > 0).all():  # if all values are positive, apply box-cox
        transformed_data = PowerTransformer(method='box-cox').fit_transform(df.iloc[:, i].values.reshape(-1, 1))
        df_boxcox_yeo_johnson[df.columns[i]] = pd.Series(transformed_data.reshape(-1))
    else:  # if any value is negative, apply default yeo-johnson
        transformed_data = PowerTransformer().fit_transform(df.iloc[:, i].values.reshape(-1, 1))
        df_boxcox_yeo_johnson[df.columns[i]] = pd.Series(transformed_data.reshape(-1))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distributions after box-cox and yeo-johnson transformation')
sns.histplot(data=df_boxcox_yeo_johnson.iloc[:, 0], ax=ax[0, 0])
sns.histplot(data=df_boxcox_yeo_johnson.iloc[:, 1], ax=ax[0, 1])
sns.histplot(data=df_boxcox_yeo_johnson.iloc[:, 2], ax=ax[1, 0])
sns.histplot(data=df_boxcox_yeo_johnson.iloc[:, 3], ax=ax[1, 1])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distributions after box-cox and yeo-johnson transformation')
qqplot(df_boxcox_yeo_johnson.iloc[:, 0], ax=ax[0, 0], line='45')
qqplot(df_boxcox_yeo_johnson.iloc[:, 1], ax=ax[0, 1], line='45')
qqplot(df_boxcox_yeo_johnson.iloc[:, 2], ax=ax[1, 0], line='45')
qqplot(df_boxcox_yeo_johnson.iloc[:, 3], ax=ax[1, 1], line='45')
plt.show()
```

```python
print('Shapiro-Wilk Test for Gaussian distribution after box-cox and yeo-johnson transformation\n')
for col in df_boxcox_yeo_johnson:
    stat, p = shapiro(df_boxcox_yeo_johnson[col])
    print(col, '\nStatistic value:', stat, '\nP-value:', p, '\n')
```
