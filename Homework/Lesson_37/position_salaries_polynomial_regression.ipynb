```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('position_salaries.csv')
df
```

```python
plt.scatter(df.Level, df.Salary)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
x, y = np.array(df.Level).reshape(-1, 1), np.array(df.Salary).reshape(-1, 1)
x_poly = PolynomialFeatures().fit_transform(x)
```

```python
lin_reg = LinearRegression().fit(x_poly, y)
```

```python
y_poly_pred = lin_reg.predict(x_poly)
```

```python
rmse = np.sqrt(mean_squared_error(y, y_poly_pred))
rmse
```

```python
lin_reg.score(x_poly, y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, label='Actual data')
plt.plot(x, y_poly_pred, color='g', label='Predicted data')
plt.xlabel('Predictor')
plt.ylabel('Target')
plt.legend()
plt.show()
```
