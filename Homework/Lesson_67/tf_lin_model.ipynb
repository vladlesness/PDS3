```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Normalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import mae
```

```python
np.random.seed(101)
```

```python
x = np.linspace(147, 210, 300)
y = np.linspace(40, 110, 300)
```

```python
x += np.random.uniform(-10, 10, 300)
y += np.random.uniform(-3, 3, 300)
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data')
plt.show()
```

```python
normalizer_layer = Normalization(input_shape=(1,), axis=None)
normalizer_layer.adapt(x)

model = Sequential([
    normalizer_layer,
    Dense(1)
])
model.compile(optimizer=Adam(learning_rate=0.1), loss=mae)
model_saved = model.fit(x, y, epochs=130, validation_split=0.2)
```

```python
plt.plot(model_saved.history['loss'],label='Train_loss')
plt.xlabel('Epochs')
plt.ylabel('Train_loss')
plt.title('Train losses over epochs')
plt.legend()
```

```python
plt.plot(model_saved.history['val_loss'],label='Val_loss')
plt.xlabel('Epochs')
plt.ylabel('Val_loss')
plt.title('Validation losses over epochs')
plt.legend()
```

```python
plt.scatter(x, y, label ='Original data')
plt.plot(x, model.predict(x), 'r', label ='Fitted line')
plt.title('Linear Regression Result')
plt.legend()
plt.show()
```
