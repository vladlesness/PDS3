```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_img(img_path):
    img = cv2.imread(img_path) 
    plt.axis('off')
    plt.imshow(img)
```

```python
show_img('cx5.png')
```

```python
def save_img(img, filename):
    cv2.imwrite(filename, img)
```

```python
img = cv2.imread('cx5.png')
```

```python
plt.imshow(img)
```

```python
def adjust_contrast_brightness(img, alpha=1.0, beta=0):
    # alpha - contrast, the default value is 1.0
    # beta - brightness, the default value is 0
    table = np.array([alpha*i + beta for i in np.arange(0, 256)]).clip(0, 255).astype('uint8')
    return cv2.LUT(img, table)
```

```python
adjusted_img = adjust_contrast_brightness(img, 1.3, -10)
```

```python
plt.imshow(adjusted_img)
```

```python
def adjust_saturation(img, satur_coef=1.0):
    img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img_HSV[:, :, 1] = img_HSV[:, :, 1] * satur_coef
    return cv2.cvtColor(img_HSV, cv2.COLOR_HSV2BGR)
```

```python
plt.imshow(adjust_saturation(img, 0.5))
```

### For changing the shape of an image cv2 has a function resize

```python
plt.imshow(cv2.resize(img, (1920, 1080)))
```

```python
def clip_img(img, x, y):
    return img[x[0]:x[1], y[0]:y[1]]
```

```python
plt.imshow(clip_img(img, (150,750), (180,1400)))
```

```python
def rotate_img(img, degree):
    degree_map = {90: 0, 180: 1, 270:2}
    return cv2.rotate(img, degree_map[degree])
```

```python
plt.imshow(rotate_img(img, 90))
```
