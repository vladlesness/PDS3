```python
import tensorflow
from tensorflow import image
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
```

```python
img = np.array(load_img('bmw.jpg'))
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
def visualize(original, augmented):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    
    ax1.imshow(original)
    ax1.set_title('Original')
    ax1.axis('off')
    
    if augmented[0][0].shape[0] == 1:
        ax2.imshow(augmented, cmap='gray')
    else:
        ax2.imshow(augmented)
    ax2.set_title('Augmented')
    ax2.axis('off')
```

```python
flipped = image.flip_left_right(img)
```

```python
visualize(img, flipped)
```

```python
saturated = image.adjust_saturation(img, 3)
```

```python
visualize(img, saturated)
```

```python
gray = image.rgb_to_grayscale(img)
```

```python
visualize(img, gray)
```

```python
brighter = image.adjust_brightness(img, 0.3)
```

```python
visualize(img, brighter)
```

```python
rotated = image.rot90(img)
```

```python
visualize(img, rotated)
```

```python
cropped = image.central_crop(img, 0.8)
```

```python
visualize(img, cropped)
```
