```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import dlib
from imutils import face_utils
```

### Метод Віоли-Джонса

```python
face_cascade = cv2.CascadeClassifier(r'D:\Python_and_DS\Startup_IT_Academy\PDS3\venv\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml')
```

```python
img = cv2.cvtColor(cv2.imread('titanic.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray_img, 1.3, 5)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 5)
```

```python
plt.imshow(img)
```

```python
rotated_img = cv2.rotate(cv2.cvtColor(cv2.imread('titanic.jpg'), cv2.COLOR_BGR2RGB), 0)
plt.imshow(rotated_img)
```

```python
rotated_gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
faces = face_cascade.detectMultiScale(rotated_gray_img, 1.3, 5)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(rotated_img, (x, y), (x+w, y+h), (255, 0, 0), 5)
```

```python
plt.imshow(rotated_img)
```

### Метод HOG

```python
img = cv2.cvtColor(cv2.imread('titanic.jpg'), cv2.COLOR_BGR2RGB)
img.astype('uint8')
```

```python
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(img, 1)

for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(img)
```

```python
rotated_img = cv2.rotate(cv2.cvtColor(cv2.imread('titanic.jpg'), cv2.COLOR_BGR2RGB), 0)
```

```python
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(rotated_img, 1)

for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(rotated_img, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(rotated_img)
```
