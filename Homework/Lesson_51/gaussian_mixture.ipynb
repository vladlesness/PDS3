```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

```python
samples, _ = make_blobs(n_samples=(8000, 5000), cluster_std=(.5, .7), center_box=(9.8, 15))
```

```python
plt.figure(figsize=(18,12))
plt.scatter(samples[:, 0], samples[:, 1])
```

```python
gm = GaussianMixture(n_components=2).fit(samples)
```

```python
plt.figure(figsize=(18,12))
plt.scatter(samples[:, 0], samples[:, 1], label='Samples')
plt.scatter(gm.means_[:, 0], gm.means_[:, 1], c='r', label='Centers')
plt.legend()
```

```python
pred = gm.predict(samples)
```

```python
df = pd.DataFrame({'x1': samples[:, 0], 'x2': samples[:, 1], 'label': pred})
groups = df.groupby('label')

fig, ax = plt.subplots()
for name, group in groups:
    ax.scatter(group.x1, group.x2, label=name)

ax.legend()
plt.show()
```
